```python
import matplotlib.pyplot as plt
import pandas as pd
from textwrap import wrap
```

```python
def get_df(filename):
    return pd.read_csv(f'{filename}.txt',
                 header=0,
                 index_col=0,
                 engine='python',
                 sep='\s*&\s*',
                 skipinitialspace=True)
```

```python
df_best_columns = pd.DataFrame([[1.69291762e+06, 3.90392940e-01, 3.81438069e-01, 1.14613331e+06]], columns=['RMSE', 'RMSLE', 'R2', 'MAE'], index=['escuelascercanas-piscina-usosmultiples-banos-habitaciones-metroscubiertos'])
```

Generar gráfico para métrica R2

```python
df = get_df('top5r2')
fig, ax = plt.subplots(figsize=(5, 11))
bars = ax.barh([ '\n'.join(wrap(l, 27)) for l in df.index.str.replace('-', ', ') ], df['R2'],height=0.5)
for bar in bars:
  width = bar.get_width() #Previously we got the height
  label_y_pos = bar.get_y() + bar.get_height() / 2
  ax.text(width, label_y_pos, s=f'{width}', va='center')
plt.xlabel("Valor de R2")
plt.ylabel("Subconjunto de columnas")
plt.title("Top 5 por R2")
plt.xlim(0.40, 0.41)
plt.show()
```
