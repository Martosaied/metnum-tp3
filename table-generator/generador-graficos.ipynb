```python
import matplotlib.pyplot as plt
import pandas as pd
from textwrap import wrap
```

```python
def get_df(filename):
    return pd.read_csv(f'{filename}.txt',
                 header=0,
                 index_col=0,
                 engine='python',
                 sep='\s*&\s*',
                 skipinitialspace=True)
```

Generar gráfico para métrica R2

```python
df = get_df('top5r2')
plt.figure(figsize=(5, 11))
plt.barh([ '\n'.join(wrap(l, 20)) for l in df.index.str.replace('-', ', ') ], df['R2'],height=0.5)
plt.xlabel("Valor de R2")
plt.ylabel("Subconjunto de columnas")
plt.title("Top 5 por R2")
plt.xlim(0.40, 0.41)
plt.show()
```
